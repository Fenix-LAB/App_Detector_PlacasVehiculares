```python
import pandas as pd
import cv2
import matplotlib.pyplot as plt

df=pd.read_csv("/content/drive/MyDrive/Red_Neuronal/Radiografias_Data_Set.csv")
```

```python
df.head(5)
```

```python
df["Finding Labels"].value_counts().to_frame()
```

```python
df_img=df.iloc[0:100,0:2]
Imagenes=[]

for name in df.iloc[0:100,0]:
  #print(name)
  dir = '/content/drive/MyDrive/Red_Neuronal/100_Images/image'
  dire = dir.replace("image",name)
  #print(dire)
  ima = cv2.imread(dire) 
  Imagenes.append(ima)
```

```python
df_test=df.iloc[100:126,0:2]
Test=[]

for name in df.iloc[100:126,0]:
  #print(name)
  dir = '/content/drive/MyDrive/Red_Neuronal/20_Images/image'
  dire = dir.replace("image",name)
  #print(dire)
  ima = cv2.imread(dire) 
  Test.append(ima)
```

```python
df_img['Image'] = Imagenes
```

```python
df_test['Image'] = Test
```

```python
df_img.head(3)
```

```python
df_img.tail(3)
```

```python
df_test.head(3)
```

```python
from tensorflow import keras

model = keras.Sequential()
    #keras.layers.Flatten(input_shape=(1024, 1024)),
model.add(keras.layers.Conv2D(32, (3, 3), activation='relu', input_shape=(1024, 1024, 3)))
model.add(keras.layers.Dense(10000, activation='relu'))
model.add(keras.layers.Dense(8))


```

```python
model.summary()
```

```python
df_test.iloc[:,2]
```

```python
labels_img=df_img['Finding Labels'].to_list()
labels_test=df_test['Finding Labels'].to_list()
```

```python
model.compile(optimizer='adam',
              loss=keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])

history = model.fit(Imagenes, labels_img, epochs=10, 
                    validation_data=(Test, labels_test))
```
